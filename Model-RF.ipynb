## Data processing

## Librerias

```python
# Bloque 1 - Importamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

## Cargamos y exploramos los datos

```python
# Bloque 2 - cargamos y exploramos los datos
melb_df = pd.read_csv("Students_Data.csv")
melb_df.index += 1
melb_df
```

## Limpiamos los datos

```python
melb_df = melb_df.dropna(axis=0)
melb_df
```

## Eliminamos las columnias de codigo de interes y aptitud (no son relevantes) + las carreras (pruebas) 

```python
# --- Asegúrate de que 'df' sea tu DataFrame principal ---

# 1. Define las columnas de código que quieres eliminar.
columnas_a_eliminar = [
    'puntaje_interes_C',
    'puntaje_interes_H',
    'puntaje_interes_A',
    'puntaje_interes_S',
    'puntaje_interes_I',
    'puntaje_interes_D',
    'puntaje_interes_E',
    
    'puntaje_aptitud_C',
    'puntaje_aptitud_H',
    'puntaje_aptitud_A',
    'puntaje_aptitud_S',
    'puntaje_aptitud_I',
    'puntaje_aptitud_D',
    'puntaje_aptitud_E',

    'perfil_interes_1_codigo', 
    'perfil_interes_2_codigo', 
    'perfil_aptitud_1_codigo', 
    'perfil_aptitud_2_codigo'
]

# 2. Verifica cuáles de esas columnas existen realmente en tu DataFrame.
columnas_existentes = [col for col in columnas_a_eliminar if col in melb_df.columns]

# 3. Si se encontraron columnas, elimínalas.
if columnas_existentes:
    print(f"Eliminando las siguientes columnas: {columnas_existentes}")
    # Usamos inplace=True para modificar 'df' directamente.
    melb_df.drop(columns=columnas_existentes, inplace=True, errors='ignore')
    print("¡Columnas eliminadas!")
    print("Nuevas dimensiones del DataFrame:", melb_df.shape)
else:
    print("Las columnas de código especificadas ya no se encuentran en el DataFrame.")
```

## Codificacion de variables objetivos Y (target)

```python

df = melb_df.copy()

# Columnas objetivos a convertir
columnas_objetivo = [
    'perfil_interes_1_nombre', 'perfil_interes_2_nombre',
    'perfil_aptitud_1_nombre', 'perfil_aptitud_2_nombre',
    'carrera_interes_1', 'carrera_interes_2', 'carrera_interes_3',
    'carrera_interes_4', 'carrera_interes_5', 'carrera_aptitud_1',
    'carrera_aptitud_2', 'carrera_aptitud_3', 'carrera_aptitud_4',
    'carrera_aptitud_5'
]

# Filtra para asegurarte de que solo usas columnas que existen
columnas_existentes = [col for col in columnas_objetivo if col in df.columns]

print(f"Columnas encontradas para codificar: {len(columnas_existentes)}")

if columnas_existentes:
    todos_los_valores = pd.concat([df[col].astype(str) for col in columnas_existentes]).unique()
    le = LabelEncoder()
    le.fit(todos_los_valores)

    # 3. Aplica la transformación usando .loc para ser explícitos
    for col in columnas_existentes:
        # Usamos .loc[:, col] para indicar "todas las filas, esta columna"
        df.loc[:, col] = le.transform(df[col].astype(str))

    print("\nCodificación completada (sin warnings). Primeras 5 filas:")
    print(df[columnas_existentes].head())

    # --- Ahora 'df' contiene los cambios. Si quieres que 'melb_df' los tenga: ---
    # melb_df = df

else:
    print("¡Advertencia! Ninguna de las columnas objetivo especificadas se encontró.")
```

## Verficacion de encoding de los terget(objetivos) la variable Y

```python
# Asegúrate de que 'le' es tu LabelEncoder ya "entrenado" (fitted)
# del paso anterior.

# Creamos un diccionario para ver fácilmente la correspondencia
mapeo_carrera_numero = dict(zip(le.classes_, le.transform(le.classes_)))

print("--- Mapeo de Carreras/Perfiles a Números ---")

# Imprimimos cada carrera/perfil y su número asignado
for nombre, numero in mapeo_carrera_numero.items():
    print(f"'{nombre}'  ->  {numero}")

print(f"\nTotal de categorías únicas codificadas: {len(mapeo_carrera_numero)}")
```

## Separacion de mis variables X (Features/Caracteristicas) Y(Target/Objetivos)

```python
# --- Asumiendo que 'df' es tu DataFrame listo y codificado ---

# 1. Define tus columnas objetivo (¡Asegúrate de que esta lista es correcta!)
columnas_objetivo = [
    'perfil_interes_1_nombre', 'perfil_interes_2_nombre',
    'perfil_aptitud_1_nombre', 'perfil_aptitud_2_nombre',
    'carrera_interes_1', 'carrera_interes_2', 'carrera_interes_3',
    'carrera_interes_4', 'carrera_interes_5', 'carrera_aptitud_1',
    'carrera_aptitud_2', 'carrera_aptitud_3', 'carrera_aptitud_4',
    'carrera_aptitud_5'
]

# Filtra solo las columnas que realmente existen
columnas_existentes = [col for col in columnas_objetivo if col in df.columns]

# 2. Separa X (Características) e Y (Objetivos)
Y = df[columnas_existentes]
X = df.drop(columns=columnas_existentes)

# 3. Divide los datos en Entrenamiento y Prueba (70/30)
#    Usamos random_state=42 para que la división sea siempre la misma.
#    Puedes usar cualquier número o quitarlo si quieres una división
#    aleatoria cada vez.
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.20, random_state=42
)

# 4. Muestra las dimensiones de los conjuntos creados
print("--- Separación y División en Entrenamiento/Prueba Completada ---")
print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de Y_train: {Y_train.shape}")
print(f"Forma de X_test:  {X_test.shape}")
print(f"Forma de Y_test:  {Y_test.shape}")

# Ahora tienes X_train, Y_train para entrenar tu modelo,
# y X_test, Y_test para evaluarlo.
```

## Por ultimo, pasamos los dtypes a int de la variable Y (en caso esten en formato Object)

```python
# 1. Verifica los tipos de datos actuales (opcional, para confirmar)
print("Tipos de datos actuales en Y_train:")
print(Y_train.dtypes)

# 2. Convierte TODAS las columnas de Y_train e Y_test a tipo entero (int)
Y_train = Y_train.astype(int)
Y_test = Y_test.astype(int)

print("\n--- ¡Y_train e Y_test convertidos a enteros! ---")

# 3. Verifica los nuevos tipos (opcional)
print("\nNuevos tipos de datos en Y_train:")
print(Y_train.dtypes)
```

# Aquí termina la etapa de DataProcessing

# Primer Modelo - Random Forest 

### Entrenamiento del modelo

```python
from sklearn.ensemble import RandomForestClassifier
import time

# --- Asumiendo que ya tienes X_train, Y_train, X_test, Y_test ---

# 1. Crea la instancia del modelo RandomForestClassifier
#    - n_estimators: El número de árboles que tendrá tu bosque. 100 es un buen punto de partida.
#    - random_state: Para que tus resultados sean reproducibles.
#    - n_jobs=-1: Usa todos los procesadores disponibles para ir más rápido (¡útil!).
modelo_rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

# 2. Entrena el modelo
print("Iniciando el entrenamiento del Random Forest...")
print("(Esto puede tardar unos momentos dependiendo de tus datos y tu PC)")

start_time = time.time() # Medimos el tiempo
modelo_rf.fit(X_train, Y_train)
end_time = time.time()

print(f"¡Entrenamiento completado en {end_time - start_time:.2f} segundos!")

# --- Tu modelo 'modelo_rf' ahora está entrenado y listo para el siguiente paso ---
```

### Pruebas de Prediccion

```python
import pandas as pd
import numpy as np
from IPython.display import display # Para mejorar la visualización en Jupyter

# --- Asumiendo que ya tienes en memoria: ---
# modelo_rf: Tu modelo Random Forest multi-salida entrenado.
# X_test: Tu conjunto de características de prueba.
# Y_test: Tu conjunto de objetivos reales de prueba.
# le: Tu LabelEncoder entrenado con todas las Y.

print("--- Realizando Predicciones con el Modelo Entrenado ---")

# 1. Usa el modelo para predecir sobre TODO el conjunto X_test
#    'Y_pred' contendrá las 4 predicciones para cada estudiante en X_test.
Y_pred = modelo_rf.predict(X_test)

print(f"¡Listo! Se generaron {Y_pred.shape[0]} predicciones con {Y_pred.shape[1]} salidas cada una.")

# 2. Vamos a mostrar un ejemplo con las 5 primeras predicciones.
X_ejemplo = X_test.head()
Y_true_ejemplo_encoded = Y_test.head()
Y_pred_ejemplo_encoded = Y_pred[:5] # Tomamos las 5 primeras filas de la predicción

# 3. Convertimos las predicciones (NumPy) a DataFrame para que se vea mejor
Y_pred_ejemplo_df = pd.DataFrame(Y_pred_ejemplo_encoded, 
                                 columns=Y_test.columns, 
                                 index=X_ejemplo.index)

print("\n--- Ejemplo: Predicción para los 5 primeros estudiantes ---")
print("\nCaracterísticas de Entrada (X_test.head()):")
display(X_ejemplo)

# 4. Intentamos traducir los números a NOMBRES para que sea más legible
try:
    # Función para aplicar inverse_transform a cada celda
    def traducir_celda(valor_numerico):
        return le.inverse_transform([valor_numerico])[0]

    # Aplicamos la traducción
    Y_true_ejemplo_nombres = Y_true_ejemplo_encoded.applymap(traducir_celda)
    Y_pred_ejemplo_nombres = Y_pred_ejemplo_df.applymap(traducir_celda)

    print("\nResultados Reales (Y_test.head()):")
    display(Y_true_ejemplo_nombres)

    print("\nResultados Predichos por el Modelo:")
    display(Y_pred_ejemplo_nombres)

except NameError:
    print("\n(No se encontró 'le', mostrando resultados codificados)")
    print("\nResultados Reales (Y_test.head() - Codificado):")
    display(Y_true_ejemplo_encoded)
    print("\nResultados Predichos por el Modelo (Codificado):")
    display(Y_pred_ejemplo_df)
except Exception as e:
    print(f"\n(Ocurrió un error al traducir: {e}. Mostrando resultados codificados)")
    print("\nResultados Reales (Y_test.head() - Codificado):")
    display(Y_true_ejemplo_encoded)
    print("\nResultados Predichos por el Modelo (Codificado):")
    display(Y_pred_ejemplo_df)
```

### Validaciones

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import numpy as np
import pandas as pd

# --- Asumiendo que YA tienes entrenado tu modelo 'modelo_rf' (con 4 salidas) ---
# --- Y que YA tienes tus conjuntos X_test, Y_test (con 4 columnas en Y_test) ---

# 1. Realiza las predicciones con tu modelo de 4 salidas
print("Realizando predicciones (4 salidas)...")
Y_pred = modelo_rf.predict(X_test)
print("¡Predicciones completadas!")

# 2. Calcula la Precisión Exacta (Manual)
#    ¿Cuántas veces acertó las 4 salidas a la vez?
exact_matches = np.all(Y_test.values == Y_pred, axis=1).sum()
total_samples = Y_test.shape[0]
exact_match_accuracy = exact_matches / total_samples

# 3. Calcula las métricas (F1, Precision, Recall) para cada salida y sus promedios
f1_list, precision_list, recall_list, accuracy_list = [], [], [], []

print("\n--- Rendimiento Individual por Salida ---")
for i, col_name in enumerate(Y_test.columns):
    y_true_col = Y_test.iloc[:, i]
    y_pred_col = Y_pred[:, i]
    
    acc = accuracy_score(y_true_col, y_pred_col)
    f1 = f1_score(y_true_col, y_pred_col, average='weighted', zero_division=0)
    prec = precision_score(y_true_col, y_pred_col, average='weighted', zero_division=0)
    rec = recall_score(y_true_col, y_pred_col, average='weighted', zero_division=0)
    
    accuracy_list.append(acc)
    f1_list.append(f1)
    precision_list.append(prec)
    recall_list.append(rec)
    
    print(f"  - {col_name}: Acc={acc:.3f}, F1={f1:.3f}, P={prec:.3f}, R={rec:.3f}")

# 4. Imprime los resultados generales
print("\n--- Métricas Generales (Promedios y Exactitud) ---")
print(f"Precisión Exacta (Exact Match - las 4 correctas): {exact_match_accuracy:.4f} ({exact_match_accuracy * 100:.2f}%)")
print(f"Accuracy Promedio (por salida):                    {np.mean(accuracy_list):.4f}")
print(f"F1-Score Promedio (Weighted, por salida):          {np.mean(f1_list):.4f}")
print(f"Precision Promedio (Weighted, por salida):       {np.mean(precision_list):.4f}")
print(f"Recall Promedio (Weighted, por salida):          {np.mean(recall_list):.4f}")
print("\nAUC: Omitido por complejidad y errores previos en formato multiclass-multioutput.")
print("--------------------------------------------------")
```
