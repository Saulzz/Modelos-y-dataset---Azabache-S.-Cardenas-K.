# DATA PROCESSING

```python
# Bloque 1 - Importamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

## Cargamos y exploramos los datos

```python
# Bloque 2 - cargamos y exploramos los datos
melb_df = pd.read_csv("Students_Data.csv")
melb_df.index += 1
melb_df
```

## LIMPIEZA DE DATOS

```python
melb_df = melb_df.dropna(axis=0)
melb_df
```

## Eliminamos las columnias de codigo de interes y aptitud (no son relevantes) + las carreras (pruebas) 

```python
# --- Asegúrate de que 'df' sea tu DataFrame principal ---

# 1. Define las columnas de código que quieres eliminar.
columnas_a_eliminar = [
    'puntaje_interes_C',
    'puntaje_interes_H',
    'puntaje_interes_A',
    'puntaje_interes_S',
    'puntaje_interes_I',
    'puntaje_interes_D',
    'puntaje_interes_E',
    
    'puntaje_aptitud_C',
    'puntaje_aptitud_H',
    'puntaje_aptitud_A',
    'puntaje_aptitud_S',
    'puntaje_aptitud_I',
    'puntaje_aptitud_D',
    'puntaje_aptitud_E',
    'perfil_interes_1_codigo', 
    'perfil_interes_2_codigo', 
    'perfil_aptitud_1_codigo', 
    'perfil_aptitud_2_codigo'
]

# 2. Verifica cuáles de esas columnas existen realmente en tu DataFrame.
columnas_existentes = [col for col in columnas_a_eliminar if col in melb_df.columns]

# 3. Si se encontraron columnas, elimínalas.
if columnas_existentes:
    print(f"Eliminando las siguientes columnas: {columnas_existentes}")
    # Usamos inplace=True para modificar 'df' directamente.
    melb_df.drop(columns=columnas_existentes, inplace=True, errors='ignore')
    print("¡Columnas eliminadas!")
    print("Nuevas dimensiones del DataFrame:", melb_df.shape)
else:
    print("Las columnas de código especificadas ya no se encuentran en el DataFrame.")
```

## Codificacion de variables objetivos Y (target)

```python
df = melb_df.copy()

# Columnas objetivos a convertir
columnas_objetivo = [
    'perfil_interes_1_nombre', 'perfil_interes_2_nombre',
    'perfil_aptitud_1_nombre', 'perfil_aptitud_2_nombre',
    'carrera_interes_1', 'carrera_interes_2', 'carrera_interes_3',
    'carrera_interes_4', 'carrera_interes_5', 'carrera_aptitud_1',
    'carrera_aptitud_2', 'carrera_aptitud_3', 'carrera_aptitud_4',
    'carrera_aptitud_5'
]

# Filtra para asegurarte de que solo usas columnas que existen
columnas_existentes = [col for col in columnas_objetivo if col in df.columns]

print(f"Columnas encontradas para codificar: {len(columnas_existentes)}")

if columnas_existentes:
    todos_los_valores = pd.concat([df[col].astype(str) for col in columnas_existentes]).unique()
    le = LabelEncoder()
    le.fit(todos_los_valores)

    # 3. Aplica la transformación usando .loc para ser explícitos
    for col in columnas_existentes:
        # Usamos .loc[:, col] para indicar "todas las filas, esta columna"
        df.loc[:, col] = le.transform(df[col].astype(str))

    print("\nCodificación completada (sin warnings). Primeras 5 filas:")
    print(df[columnas_existentes].head())

    # --- Ahora 'df' contiene los cambios. Si quieres que 'melb_df' los tenga: ---
    # melb_df = df

else:
    print("¡Advertencia! Ninguna de las columnas objetivo especificadas se encontró.")
```

## Verficacion de encoding de los terget(objetivos) la variable Y

```python
# Asegúrate de que 'le' es tu LabelEncoder ya "entrenado" (fitted)
# del paso anterior.

# Creamos un diccionario para ver fácilmente la correspondencia
mapeo_carrera_numero = dict(zip(le.classes_, le.transform(le.classes_)))

print("--- Mapeo de Carreras/Perfiles a Números ---")

# Imprimimos cada carrera/perfil y su número asignado
for nombre, numero in mapeo_carrera_numero.items():
    print(f"'{nombre}'  ->  {numero}")

print(f"\nTotal de categorías únicas codificadas: {len(mapeo_carrera_numero)}")
```

## Separacion de mis variables X (Features/Caracteristicas) Y(Target/Objetivos)

```python
# --- Asumiendo que 'df' es tu DataFrame listo y codificado ---

# 1. Define tus columnas objetivo (¡Asegúrate de que esta lista es correcta!)
columnas_objetivo = [
    'perfil_interes_1_nombre', 'perfil_interes_2_nombre',
    'perfil_aptitud_1_nombre', 'perfil_aptitud_2_nombre',
    'carrera_interes_1', 'carrera_interes_2', 'carrera_interes_3',
    'carrera_interes_4', 'carrera_interes_5', 'carrera_aptitud_1',
    'carrera_aptitud_2', 'carrera_aptitud_3', 'carrera_aptitud_4',
    'carrera_aptitud_5'
]

# Filtra solo las columnas que realmente existen
columnas_existentes = [col for col in columnas_objetivo if col in df.columns]

# 2. Separa X (Características) e Y (Objetivos)
Y = df[columnas_existentes]
X = df.drop(columns=columnas_existentes)

# 3. Divide los datos en Entrenamiento y Prueba (80/20)
#    Usamos random_state=42 para que la división sea siempre la misma.
#    Puedes usar cualquier número o quitarlo si quieres una división
#    aleatoria cada vez.
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.20, random_state=42
)

# 4. Muestra las dimensiones de los conjuntos creados
print("--- Separación y División en Entrenamiento/Prueba Completada ---")
print(f"Forma de X_train: {X_train.shape}")
print(f"Forma de Y_train: {Y_train.shape}")
print(f"Forma de X_test:  {X_test.shape}")
print(f"Forma de Y_test:  {Y_test.shape}")

# Ahora tienes X_train, Y_train para entrenar tu modelo,
# y X_test, Y_test para evaluarlo.
```

## Por ultimo, pasamos los dtypes a int de la variable Y (en caso esten en formato Object)

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Crear copias para evitar SettingWithCopyWarning
Y_train_encoded = Y_train.copy()
Y_test_encoded = Y_test.copy()

# Crear un diccionario para almacenar los encoders por columna
encoders = {}

print("Codificando Y_train e Y_test columna por columna...")

# Iterar sobre cada columna objetivo
for col in Y_train.columns:
    # Crear e instanciar un nuevo LabelEncoder para esta columna
    le = LabelEncoder()

    # Ajustar y transformar Y_train para esta columna
    Y_train_encoded[col] = le.fit_transform(Y_train[col].astype(str))

    # Transformar Y_test usando el MISMO encoder ajustado
    # Manejar posibles etiquetas no vistas en Y_test (aunque raro con split)
    # Una forma es agregar las no vistas al encoder, pero es más simple
    # asegurarse de que Y_test solo contenga valores vistos o transformar
    # con cuidado. Aquí asumimos que Y_test tendrá valores vistos.
    # Para mayor seguridad, filtramos Y_test para solo incluir clases conocidas.
    known_labels = le.classes_
    Y_test_encoded[col] = Y_test[col].astype(str).apply(lambda x: le.transform([x])[0] if x in known_labels else -1) # -1 o algún valor para no vistos

    encoders[col] = le

print("\\n--- ¡Y_train e Y_test codificados por columna! ---")

# Verificar los tipos de datos (deberían ser enteros)
print("\\nNuevos tipos de datos en Y_train_encoded:")
print(Y_train_encoded.dtypes)

print("\\nPrimeras filas de Y_train_encoded:")
print(Y_train_encoded.head())
```

# Aquí termina la etapa de DataProcessing

## MODELO XGBOOST

```python
from sklearn.multioutput import MultiOutputClassifier
import xgboost as xgb
import time 

# Definimos el modelo base XGBoost
xgb_model = xgb.XGBClassifier(
    objective='multi:softprob',
    use_label_encoder=False, 
    eval_metric='mlogloss',
    random_state=42,
    verbosity=0 
)
model = MultiOutputClassifier(xgb_model, n_jobs=-1)

print("\nEntrenando el modelo XGBoost...")

start_time = time.time()
model.fit(X_train, Y_train_encoded)
end_time = time.time()

print(f"¡Entrenamiento completado en {end_time - start_time:.2f} segundos!")
```

## PREDICCIÓN DE MÉTRICAS

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from IPython.display import display

Y_train_encoded = Y_train.copy()
Y_test_encoded = Y_test.copy()
encoders = {}

print("Codificando Y_train e Y_test columna por columna...")
for col in Y_train.columns:
    le = LabelEncoder()
    y_train_str = Y_train[col].astype(str)
    y_test_str = Y_test[col].astype(str)
    Y_train_encoded[col] = le.fit_transform(y_train_str)
    known_labels = set(le.classes_)
    Y_test_encoded[col] = y_test_str.apply(lambda x: le.transform([x])[0] if x in known_labels else -1)
    encoders[col] = le
print("¡Codificación por columna completada!")


# --- Cálculo de Métricas ---
target_columns = Y_test_encoded.columns.tolist()
Y_test_np = Y_test_encoded.values
metrics_dict = {}
Y_pred = model.predict(X_test)

print("\n--- Calculando Métricas para XGBoost ---")
n_outputs = Y_test_np.shape[1]

for i in range(n_outputs):
    y_true_col = Y_test_np[:, i]
    y_pred_col = Y_pred[:, i]
    col_name = target_columns[i]

    # Filtrar valores -1 si existen (etiquetas no vistas)
    mask = y_true_col != -1
    y_true_filtered = y_true_col[mask]
    y_pred_filtered = y_pred_col[mask]

    if len(y_true_filtered) > 0:
        acc = accuracy_score(y_true_filtered, y_pred_filtered)
        pre = precision_score(y_true_filtered, y_pred_filtered, average='macro', zero_division=0)
        rec = recall_score(y_true_filtered, y_pred_filtered, average='macro', zero_division=0)
        f1 = f1_score(y_true_filtered, y_pred_filtered, average='macro', zero_division=0)
    else:
        acc, pre, rec, f1 = 0, 0, 0, 0

    metrics_dict[col_name] = {
        'Accuracy': acc,
        'Precision': pre,
        'Recall': rec,
        'F1-Score': f1
    }

print("¡Métricas calculadas!")

metrics_df = pd.DataFrame.from_dict(metrics_dict, orient='index')
avg_accuracy = metrics_df['Accuracy'].mean()
avg_precision = metrics_df['Precision'].mean()
avg_recall = metrics_df['Recall'].mean()
avg_f1 = metrics_df['F1-Score'].mean()
metrics_df.loc['Promedio General'] = [avg_accuracy, avg_precision, avg_recall, avg_f1]

print("\n======================================")
print("---   Tabla Resumen de Métricas XGBoost ---")
print("======================================")

# Usar display si está disponible, si no, print
try:
    display(metrics_df.style.format("{:.4f}"))
except NameError:
    print(metrics_df.to_string(float_format="{:.4f}".format))

print("\n======================================")
```
